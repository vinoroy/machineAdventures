```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
doorImg_fileName = '../images/door2.jpeg'
lockTempImg_fileName = '../images/door2Lock_temp.jpeg'
```

```python
# load the lock and the template lock image
lockImg = cv2.imread(doorImg_fileName,0)
tempImg = cv2.imread(lockTempImg_fileName,0)

```

```python
# show the two images
plt.subplot(121),plt.imshow(lockImg)
#plt.subplot(122),plt.imshow(cv2.cvtColor(tempImg,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
res = cv2.matchTemplate(lockImg,tempImg,eval('cv2.TM_SQDIFF_NORMED'))
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)


w, h = tempImg.shape[::-1]


top_left = min_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
```

```python
y = top_left[1]
x = top_left[0]

print y
print x


cropImg = lockImg[y:y+h, x:x+w]

plt.subplot(122),plt.imshow(cropImg)

plt.show()
```
