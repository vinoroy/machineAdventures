```python
# load the datasets
data = pd.read_csv('../data/abalone.data',header=None)
```

```python
data.head()
```

```python
data.columns = ['Sex','Length','Diam','Height','Whole','Shucked','Viscera','Shell','Rings']
```

```python
data.head()
```

```python
sns.pairplot(data.iloc[:,1:9])
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
X = data.iloc[:,0:8].values
y = data.iloc[:,8].values
X.shape
```

```python
enc = OneHotEncoder(handle_unknown='ignore')

Sex = enc.fit_transform(X[:,0].reshape(-1, 1)).toarray()

Sex = Sex[:,0:2]
```

```python
X = np.concatenate((Sex,X[:,1:]),axis=1)
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
lmodel = LinearRegression()
```

```python
lmodel.fit(X_train,y_train)
```

```python
y_pred = lmodel.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred) 
```

```python
lin_clf = svm.LinearSVC()
```

```python
lin_clf.fit(X_train, y_train)
```

```python
y_pred = lin_clf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```
